### Crimes Across the Years:

- Is the total number of crimes increasing or decreasing across the years?

- Are there any individual crimes that are doing the opposite (e.g decreasing when overall crime is increasing or vice-versa)?

```python
plot_df = df.loc['2001':'2023','Primary Type']
plot_df
```

```python
ts_y_res = df_ts.resample("Y").count()
ts_y_res.name = 'Resampled'
ts_y_res.head()
```

```python
# make the fig/axis

fig, ax = plt.subplots(figsize=(12,3))

ax.set(title='Crimes in Chicago', ylabel='Number of Crimes')


# customize major ticks every year
major_loc =  mdates.MonthLocator(interval = 1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks every 3 months no label
minor_loc =  mdates.MonthLocator(interval=3)
ax.xaxis.set_minor_locator(minor_loc)

# plot time series with Pandas
ts_y_res['Primary Type'].plot(style='.-', ax=ax, label = 'Count')


# Final xdate adjustment
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)


ax.legend();
```

There is a clear trend in decreasing crimes across the years until 2020.  For the two full years after 2020 there is a slight overall increase, and the year 2023 is not complete.

```python
## visualizing 

sns.countplot(data=df, x='Month',palette='dark');
```

```python
# Extracting the Month name

df['Month'] = df.index.month
df['MonthName'] = df.index.month_name()
df.head(3)
```

```python
# saving month name

unique_months = df.drop_duplicates(subset=['Month']).sort_values('Month')
month_lookup =dict(zip(unique_months['Month'],unique_months['MonthName']))
month_lookup
```

```python
sns.countplot(data=df, x='MonthName',palette='dark',order=month_lookup.values());
```

```python
# save month counts

month_counts = df['MonthName'].value_counts()[month_lookup.values()]
ax = month_counts.plot(style='o-',grid=True,
                       ylabel='# of Crimes',
                       xlabel='Month');
```

July and August have the most crime. February, November, and December have the least crime.
